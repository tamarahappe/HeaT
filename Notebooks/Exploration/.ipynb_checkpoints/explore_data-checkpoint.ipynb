## NASA yearly GMST

```python
yearly_GMST = pd.read_csv(f"{data_path}NASA_GMST.csv", delimiter=";", skiprows=1, index_col=0)
```

```python
yearly_GMST.plot()
```

### convert to xarray to have same shape as t2m

```python
## this is from 1940-2023, JJA (92 days)
LOWESS5y_yearly_GMST = yearly_GMST["Lowess(5)"][-84:].to_numpy() ## select 1940-2023

GMST_reshaped = np.empty(7728,) #empty array with shape of number of days (e.g 92 per summer, per year)

for i, GMST in enumerate(LOWESS5y_yearly_GMST): ## loop over the GMST data to fill the array
    year_array = np.array(92 * [GMST], dtype=float)
    GMST_reshaped[i*92:(i+1)*92] = year_array[:]
    
#create xarray 
GMST_xr = xr.DataArray(
    data=GMST_reshaped,
    dims=["time"],
    coords=dict(
        time=t2m.time.dt.year,
    ),
    attrs=dict(
        description="Global Mean Surface Temperature anomalie from NASA, 5 year LOWESS",
        units="degC",
    ),
)
```

## June and sept and selecting subset

```python
T2M_file = "1940-2023_T2M_westEU_JJAS_LENTISGRID.nc"
t2m = xr.open_dataset(f"{data_path}{T2M_file}")
```

```python
t2m.time
```

```python
june = t2m.sel(time=t2m.time.dt.month==6)
june =june.sel(time=june.time.dt.day>22)

julyaugust = t2m.sel(time=t2m.time.dt.month.isin([7,8]))

sept = t2m.sel(time=t2m.time.dt.month==9)
sept = sept.sel(time=sept.time.dt.day<8)
```

```python
june23_sept7 = xr.merge([june, julyaugust, sept])
```

```python
june23_sept7.mean(dim="time")["t2m"].plot(cmap="coolwarm")
```

```python
june23_sept7["t2m"].sel(lat=50, lon=5, method="nearest").plot()
```

```python
june23_sept7.to_netcdf(f"{data_path}1940-2023_T2M_westEU_June23-JA-Sept7_LENTISGRID.nc")
```
